```python
import matplotlib.pyplot as plt
import numpy as np
data = np.loadtxt('populations.txt')
data
```

```python
# Nomber 1

dataT = data.T

plt.grid(color = 'grey')
plt.plot(dataT[0], dataT[1], color = 'g')
plt.plot(dataT[0], dataT[2], color = 'orange')
plt.plot(dataT[0], dataT[3], color = 'b')
plt.xlabel('Years')
plt.ylabel('Populations')
plt.text(1914, 75000, 'Hare', color = 'g')
plt.text(1908, 4000, 'Lynx', color = 'orange')  
plt.text(1907, 50000, 'Carrot', color = 'b')
```

```python
# Nomber 2

plt.grid(color = 'grey')
plt.plot(dataT[0], dataT[1]/(dataT[1] + dataT[2]) * 100, color = 'g')
plt.plot(dataT[0], dataT[2]/(dataT[1] + dataT[2]) * 100, color = 'orange')
plt.text(1911, 75, 'Hare', color = 'g')
plt.text(1911, 22, 'Lynx', color = 'orange') 
plt.xlabel('Years')
plt.ylabel('Percent %')
```

```python
 # Number 3

hare_mean = np.mean(dataT[1])
hare_std = np.std(dataT[1])

lynx_mean = np.mean(dataT[2])
lynx_std = np.std(dataT[2])

carrot_mean = np.mean(dataT[3])
carrot_std = np.std(dataT[3])

print('Hare mean = ', hare_mean, ';   Hare std = ', hare_std)
print('Lynx mean = ', lynx_mean, ';   Lynx std = ', lynx_std)
print('Carrot mean = ', carrot_mean, ';   Carrot std = ', carrot_std)


fig = plt.figure(figsize = (15,5))

plt.subplot(131)
plt.grid(color = 'grey')
plt.title('Hare Population')
plt.xlabel('Years')
plt.hist(dataT[1], color = 'g', bins = 20)
plt.axvline(hare_mean, linestyle = 'dashed')

plt.subplot(132)
plt.grid(color = 'grey')
plt.title('Lynx Population')
plt.xlabel('Years')
plt.hist(dataT[2], color = 'orange', bins = 20)
plt.axvline(hare_mean, linestyle = 'dashed')

plt.subplot(133)
plt.grid(color = 'grey')
plt.title('Carrot')
plt.xlabel('Years')
plt.hist(dataT[1], color = 'b', bins = 20)
plt.axvline(hare_mean, linestyle = 'dashed')
```

```python
# Number 4

plt.grid(color = 'grey')
plt.plot(dataT[0], dataT[1], color = 'g')
plt.plot(dataT[0], dataT[2], color = 'orange')
plt.plot(dataT[0], dataT[3], color = 'b')
plt.xlabel('Years')
plt.ylabel('Populations')
plt.text(1914, 75000, 'Hare', color = 'g')
plt.text(1908, 4000, 'Lynx', color = 'orange')  
plt.text(1907, 50000, 'Carrot', color = 'b')

plt.scatter(dataT[0][np.argmax(dataT[1])], np.max(dataT[1]), color = 'g')
plt.scatter(dataT[0][np.argmax(dataT[2])], np.max(dataT[2]), color = 'orange')
plt.scatter(dataT[0][np.argmax(dataT[3])], np.max(dataT[3]), color = 'b')
            
```

```python
# Number 6

a = dataT[0][dataT[1]>50000]
b = dataT[0][dataT[2]>50000]
c = dataT[0][dataT[3]>50000]
print('Years when hares > 50000', a)
print('Years when lynx > 50000', b)
print('Years when carrots > 50000', c)

plt.grid(color = 'grey')
plt.plot(dataT[0], dataT[1], color = 'g')
plt.plot(dataT[0], dataT[2], color = 'orange')
plt.plot(dataT[0], dataT[3], color = 'b')
plt.axhline(y=50000, color='r', linestyle='-')
plt.xlabel('Years')
plt.ylabel('Populations')
plt.text(1914, 75000, 'Hare', color = 'g')
plt.text(1908, 4000, 'Lynx', color = 'orange')  
plt.text(1907, 35000, 'Carrot', color = 'b')

```

```python
# Number 7

sortedHares = sorted(dataT[1])
sortedLynx = sorted(dataT[2])

H0 = dataT[0][list(dataT[1]).index(sortedHares[0])]
H1 = dataT[0][list(dataT[1]).index(sortedHares[1])]

L0 = dataT[0][list(dataT[2]).index(sortedLynx[0])]
L1 = dataT[0][list(dataT[2]).index(sortedLynx[1])]

print('The lowest population of hares was in', H0 , 'and', H1, 'years.')
print('The lowest population of lynx was in', H0 , 'and', H1, 'years.')

```

```python
# Number 8

v2 = dataT[1][1:]
v1 = dataT[1][:20]
hares_changes = (v2 - v1) / v1 * 100

u2 = dataT[2][1:]
u1 = dataT[2][:20]
lynx_changes = (u2 - u1) / u1 * 100

fig = plt.figure(figsize = (15,5))

plt.subplot(121)
plt.grid(color = 'grey')
plt.title('Hare Population')
plt.xlabel('Years')
plt.plot(dataT[0][1:], hares_changes, color = 'g')


plt.subplot(122)
plt.grid(color = 'grey')
plt.title('Lynx Population')
plt.xlabel('Years')
plt.plot(dataT[0][1:], lynx_changes, color = 'orange')

from scipy.stats.stats import pearsonr
corr1 = np.corrcoef(hares_changes, dataT[2][1:])[0][1]
corr2 = np.corrcoef(lynx_changes, dataT[1][1:])[0][1]
print('Correlation between lynx population and changes of hares = ', corr1)
print('Correlation between hares population and changes of lynx = ', corr2)
```
